```python
def null_percentage(column):
    total = len(column)
    nulls = column.isnull().sum()
    percentage = (nulls / total) * 100
    return percentage
```

```python
def unique_values_for_int_columns(dfi1):
    unique_values = {}
    for column in dfi1.select_dtypes(include='int64').columns:
        unique_values[column] = dfi1[column].unique()
    return unique_values
```

```python
# Function to exclude empty or all-NA columns
def exclude_empty_or_all_na_columns(df):
    return df.dropna(axis=1, how='all')
```

```python
dfM1=exclude_empty_or_all_na_columns(dfM1)
dfM2=exclude_empty_or_all_na_columns(dfM2)
dfM3=exclude_empty_or_all_na_columns(dfM3)
```

```python
pd.concat([dfM1, dfM2, dfM3])
```

```python
mobi_data=pd.concat([dfM1, dfM2, dfM3])
```

```python
##merge the below 2 datasets
##1.incident_data    2. mobi_data
df=pd.merge(incident_data, mobi_data,  on=['IncidentNumber',"CalYear", "HourOfCall"],how="inner")
```

```python
df.info()
```

```python
##DATAVIZ##

#Création de la variable du temps de réponse totale
df["TotalResponseTime"] = df["TurnoutTimeSeconds"] + df["TravelTimeSeconds"] + df["AttendanceTimeSeconds"]
```

```python
plt.figure(figsize=(14, 7))
sns.boxplot(data=df, x="IncGeo_BoroughName", y="TotalResponseTime")
plt.xticks(rotation=90)
plt.title("Distribution des temps d\'intervention par région")
plt.xlabel("Région")
plt.ylabel("Temps d\'intervention (secondes)")
plt.show()
```

```python
Q1= df["TotalResponseTime"].quantile(0.25)
Q3= df["TotalResponseTime"].quantile(0.75)
```

```python
IQR=Q3-Q1
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3-1.5*IQR
```

```python
df_nooutlier=df[(df['TotalResponseTime']>=lower_limit)&
(df['TotalResponseTime']<=upper_limit)]
```

```python
plt.figure(figsize=(14, 7))
sns.boxplot(data=df_nooutlier, x="IncGeo_BoroughName", y="TotalResponseTime")
plt.xticks(rotation=90)
plt.title("Distribution des temps d\'intervention par région")
plt.xlabel("Région")
plt.ylabel("Temps d\'intervention (secondes)")
plt.show()
```

```python
##Extract Heure, Jour, et Mois de vaiable DateAndTimeMobilised
df['HeuredeMobilisation'] = df['DateAndTimeMobilised'].dt.hour
df['Jourdelasemaine'] = df['DateAndTimeMobilised'].dt.dayofweek  # Monday=0, Sunday=6
df['Mois'] = df['DateAndTimeMobilised'].dt.month
```

```python
# pour convertir inf au NAN
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
#Distribution de Response Times:
plt.figure(figsize=(12, 6)) 
sns.histplot(df["TotalResponseTime"], bins=50, kde=True) 
plt.title('Distribution de temps de réponse')
plt.xlabel('Response Time (seconds)') 
plt.ylabel('Frequency') 
plt.show()
```

```python
##TEMPS DE REPONSE EN FONCTION DE JOUR DE LA SEMAINE:
jour_labels = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']
plt.figure(figsize=(12, 6))
sns.boxplot(x='Jourdelasemaine', y="TotalResponseTime", data=df) 
plt.title('Temps de réponse en fonction de jour de la semaine') 
plt.xticks(ticks=range(7), labels=jour_labels)
plt.xlabel('Jour de la semaine')
plt.ylabel('Temps de réponse (seconds)') 
plt.show()
```

```python
plt.figure(figsize=(12, 6)) 
sns.boxplot(x='Mois', y="TotalResponseTime", data=df) 
plt.title('Temps de réponse en fonction de Mois') 
plt.xlabel('Mois') 
plt.ylabel('Temps de réponse') 
plt.show()
```

```python
##Response Time vs. Hour of Mobilisation:
plt.figure(figsize=(12, 6)) 
sns.boxplot(x='HeuredeMobilisation', y='ResponseTimeSecondsG', data=df) 
plt.title('Temps de réponse vs. Heure d mobilisation') 
plt.xlabel('Heure de mobilisation')
plt.ylabel('Temps de réponse (seconds)') 
plt.show()

```
